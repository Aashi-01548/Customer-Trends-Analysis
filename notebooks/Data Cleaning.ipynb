```python
# Customer Trends Analysis - Data Cleaning
# Portfolio Project for 4GB RAM Laptop

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
print("\n" + "=" * 60)
print("STEP 3: DATA CLEANING")
print("=" * 60)
```

```python
# Create a copy for cleaning
df_clean = df.copy()
```

```python
# 1. Remove rows with missing CustomerID
print(f"\nRows before removing missing CustomerID: {len(df_clean)}")
df_clean = df_clean.dropna(subset=['CustomerID'])
print(f"Rows after removing missing CustomerID: {len(df_clean)}")
```

```python
# 2. Remove rows with missing Description
df_clean = df_clean.dropna(subset=['Description'])
```

```python
# 3. Remove cancelled orders (InvoiceNo starting with 'C')
df_clean = df_clean[~df_clean['InvoiceNo'].astype(str).str.startswith('C')]
```

```python
# 4. Remove rows with negative or zero Quantity and UnitPrice
df_clean = df_clean[(df_clean['Quantity'] > 0) & (df_clean['UnitPrice'] > 0)]
```

```python
# 5. Convert InvoiceDate to datetime
df_clean['InvoiceDate'] = pd.to_datetime(df_clean['InvoiceDate'])
```

```python
# 6. Create new features
df_clean['TotalAmount'] = df_clean['Quantity'] * df_clean['UnitPrice']
df_clean['Year'] = df_clean['InvoiceDate'].dt.year
df_clean['Month'] = df_clean['InvoiceDate'].dt.month
df_clean['DayOfWeek'] = df_clean['InvoiceDate'].dt.day_name()
df_clean['Hour'] = df_clean['InvoiceDate'].dt.hour
```

```python
print(f"\nFinal cleaned dataset shape: {df_clean.shape}")
print(f"\nNew columns added:")
print(df_clean[['TotalAmount', 'Year', 'Month', 'DayOfWeek', 'Hour']].head())
```

```python
print("\n" + "=" * 60)
print("STEP 4: DATA QUALITY CHECKS")
print("=" * 60)
```

```python
print("\nMissing values after cleaning:")
print(df_clean.isnull().sum())
```
